```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_file(filename):
    return pd.read_csv(filename).values

#read_file("data/petrol_consumption.csv")
```

```python
def normalize_data(data):
    mean = np.mean(data, axis=0)
    st = np.std(data, axis=0)
    return (data - mean)/st 

data = read_file("data/petrol_consumption.csv")
normalize_data(data)    
```

```python
def create_training_test(data):
    col = data.shape[1]-1
    x_train = data[:int((70*len(data)) / 100), :col]
    y_train = data[:int((70*len(data)) / 100), col]
    x_test = data[int((70*len(data)) / 100):, :col]
    y_test = data[int((70*len(data)) / 100):, col]
    return x_train, y_train, x_test, y_test

data = read_file("data/petrol_consumption.csv")
#data = read_file("data/ex1data2(Home_1f).csv")
data1 = normalize_data(data)
create_training_test(data1)
```

```python
def calculate_cost( X, y, theta):
    m = y.shape[0]
    pred = X.dot(theta)
    cost = (1/(2*m))* np.sum(np.square(pred-y))
    return cost

def gradient_descent( X, y, theta, nb_iterations, learning_rate):
    '''Return the final theta vector and array of cost history over nro of iterations'''
    print("nb_it")
    print(nb_iterations)
    m = y.shape[0]
    cost_history = np.zeros(nb_iterations)
    for i in range(nb_iterations):
        prediction = X.dot(theta)
        theta = theta- ((1/m) * learning_rate * (X.T.dot(prediction - y)))
        cost_history[i] = calculate_cost( X, y, theta) 
    return theta, cost_history

data = read_file("data/petrol_consumption.csv")
#data = read_file("data/ex1data2(Home_1f).csv")
data = normalize_data(data)
x_train, y_train, x_test, y_test = create_training_test(data)
#theta = np.random.randn (1,1)
theta = np.random.randn(4,1) #"data/petrol_consumption.csv"

nb_iterations=100
learning_rate=0.01
theta, cost_history = gradient_descent(x_train, y_train, theta, nb_iterations, learning_rate)
#print(theta)
#print(theta_history)
#print(cost_history)
```

```python
def normal_equation(X, y):
    X_t = np.transpose(X)
    return np.matmul(np.linalg.inv(np.matmul(X_t, X)), np.matmul(X_t, y))
    #theta = np.linalg.inv(X.T.dot(X)).dot((X.T).dot(y))
    #return theta
```

```python
def create_theta(X):
    return np.random.rand(X.shape[1])
```
